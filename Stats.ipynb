```python
data_nd = sns.load_dataset('titanic')
data_nd.sample(7)
```

```python
plt.figure(figsize=(5,3))
data_nd['age'].plot(kind='kde')
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x = data_nd['age'])
```

```python
# z_score -> scaling down the data
s = []
for i in data_nd['age']:
  s.append((i - data_nd['age'].mean()) / data_nd['age'].std()) # z_score formula
s
```

```python
z_score_data = pd.DataFrame({'age':s})
z_score_data
```

```python
plt.figure(figsize=(5,3))
z_score_data['age'].plot(kind='kde')
```

**Covariance**

covariance is used to find the relation between 2 different columns.

covariance formula = summation(x - x_mean)*(y - y_mean) / (n-1)

covariance range = (-inf to inf)

```python
df = sns.load_dataset('tips')
df.sample(7)
```

```python
df.shape
```

```python
df.info()
```

```python
df_n = df.select_dtypes(exclude='category')
df_n.head(7)
```

```python
df_n.cov()
```

**correlation**

correlation will work with Numerical Columns only.

correlation will find both relation and strength.

range of correlation = -1 to 1

-1 to 0 -> negative relation

0 -> no relation

0 to 1 -> positive relation

* There are 2 types of Maths
  
  * 1.Pearson Correlation
  * 2.SpearMan Rank Correlation

```python
df_n.head(5)
```

```python
df_n.corr(method='pearson')
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(df_n.corr(method='pearson'),vmin=-1,vmax=1,annot=True,cmap='tab20')
```

```python
# spearman rank correlation
```

```python
df_n.corr(method='spearman')
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(df_n.corr(method='spearman'),vmin=-1,vmax=1,annot=True,cmap='tab20')
```
